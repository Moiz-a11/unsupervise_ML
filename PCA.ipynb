```python
from sklearn.datasets import  load_iris
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X=pd.DataFrame(iris.data)
y=iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#PCA
from sklearn.decomposition import PCA
pca  = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled) # successfully converted 4D into 2d
```

```python
print("varience of f1 and f2" ,pca.explained_variance_ratio_)
```

```python
print(pca.components_())#eigen vectors
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=y
)
plt.xlabel("pc line 1")
plt.ylabel("pc line 2")
plt.title("PCA for iris dataset")
```
